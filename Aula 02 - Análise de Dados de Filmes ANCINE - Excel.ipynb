# Análise de Dados de Filmes Exibidos no Brasil de 2009 a 2017

Com base nos dados da Ancine (https://oca.ancine.gov.br/) vamos analisar os filmes exibidos no Brasil

Arquivo -> 2120_0.xlsx

```python
# importando bibliotecas
import pandas as pd
import os

# Informar ao matplotlib sobre gráficos inline
%matplotlib inline
```

```python
# Criar o DataFrame com base no arquivo Excel
# Precisamos pular as duas primeiras linhas
df = pd.read_excel('data/2120_0.xlsx', skiprows=2, skip_footer=16)

df.head()
```

```python
# Exibir as 5 ultimas linhas
df.tail(20)
```

```python
# Verificar todas os tipos das colunas
df.dtypes
```

```python
# Vamos tentar identificar o problema da coluna Data de lançamento
df['Data de lançamento'].value_counts()
```

```python
# Exibir alguns filmes que estão com data = Relançamento
df[df['Data de lançamento'] == 'Relançamento'].head()
```

## Decisão sobre dados incorretos/faltantes/divergentes

```python
df.shape
```

```python
# Criar um novo DataFrame sem os Relançamentos
# copy - para não cria uma cópia do zero. 
# Sem copy, faz referência a um espaço de memoria, em vez de criar do zero
df_novo = df[df['Data de lançamento'] != 'Relançamento'].copy()
```

```python
# Mostrar o tamanho do DataFrame
df_novo.shape
```

```python
df_novo.dtypes
```

```python
# Converter o campo Data de lançamento em Data
df_novo['Data de lançamento'] = pd.to_datetime(df_novo['Data de lançamento'])
```

```python
df_novo.dtypes
```

```python
# Quais os anos tiveram mais filmes lançados?
df_novo['Data de lançamento'].dt.year.value_counts()
```

```python
# Vamos plotar um gráficos com os lançamentos por ano
df_novo['Data de lançamento'].dt.year.value_counts().plot.bar()
```

```python
# plotar gráfico de pizza de genero dos filmes
df_novo['Gênero'].value_counts().plot.pie()
```

```python
# Qual filme com a maior biliteria?
df_novo[df_novo['Renda (R$) no ano de exibição'] == df_novo['Renda (R$) no ano de exibição'].max()]
```

```python
# Qual filme com a maior biliteria?
df_novo[df_novo['Renda (R$) no ano de exibição'] == df_novo['Renda (R$) no ano de exibição'].min()].head()
```

```python
# #filmes com mairo bilheteria
df_novo.nlargest(3, 'Renda (R$) no ano de exibição')
```

```python
# Quantos filmes brasileiros e estrageiros?
df_novo['Nacionalidade da obra'].value_counts()
```
